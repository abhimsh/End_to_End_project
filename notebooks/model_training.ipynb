```python
diamond_data.drop("id", axis=1, inplace=True)
```

```python
X = diamond_data.iloc[:, :-1]
Y = diamond_data.iloc[:, -1:]
```

```python
categorical_columns = X.columns[X.dtypes == "object"]
numerical_columns = X.columns[X.dtypes != "object"]
print(categorical_columns)
print(numerical_columns)
```

```python
X.select_dtypes(include="object")
```

```python
X.select_dtypes(exclude="object")
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler

# pipelines
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=52)
```

```python
num_pipeline = Pipeline(
    steps=[
        ("imputer", SimpleImputer()),
        ("standardization", StandardScaler())
    ]
)

order_categories = [
    ["Fair", "Good", "Very Good", "Premium" , "Ideal"],
    ["D", "E", "F", "G", "H", "I", "J"], 
    ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
    ]

cat_pipeline = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("encoding", OrdinalEncoder(categories=order_categories))
    ]
)
```

```python
preprocessor = ColumnTransformer(
    [
        ("num_pipeline", num_pipeline, numerical_columns),
        ("cat_pipeline", cat_pipeline, categorical_columns)
    ]
)
```

```python
X_train
```

```python
y_train
```

```python
preprocessor.fit_transform(X_train)
```

```python
preprocessor.transform(X_test)
```

```python
preprocessor.get_feature_names_out()
```

```python
X_train = pd.DataFrame(preprocessor.fit_transform(X_train), columns=preprocessor.get_feature_names_out())
X_test = pd.DataFrame(preprocessor.transform(X_test), columns=preprocessor.get_feature_names_out())
```
